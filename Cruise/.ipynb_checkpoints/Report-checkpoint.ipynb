Table schema description:

user_id: identification number assigned to each user

car_id: name of each car used for the ride

start_time: start time of the ride in YYYY-MM-DD HH:MM:SS

end_time: end time of the ride in YYYY-MM-DD HH:MM:SS

num_riders: number of passengers in the ride

region: the region that the car is operating in (origin)

num_near_misses: the number of times that the car is close to getting into an accident during the ride

price: the price that the rider paid to get on the ride

rating: the rating (1 through 5) that the rider gave the ride (5 is best)



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("ggplot")
%matplotlib inline
```

```python
data['start_time'] = pd.to_datetime(data['start_time'], format="%Y-%m-%d %H:%M:%S")
data['end_time'] = pd.to_datetime(data['end_time'], format="%Y-%m-%d %H:%M:%S")
data['duration_second'] = (data['end_time'] - data['start_time']).astype('timedelta64[s]')
```

```python
data.describe(include = 'all')
```

Ride duartion has negative values, which means the end time is earlier than start time. From the above description table start and end time columns, we can see the data contains one day worth of data (2018-10-02). Let's see these subset of data which contains the anomaly.

```python
data[data.duration_second<0]
```

For example row 1032, it start at 2018-10-02 23:59, and ends at the same day but in the morning. These could be some data ingesting issue that the date part remains the same, while actually it should be the next day early morning. To fix this we can: when the duration is negative, add 1 day to end date.

```python
data.loc[data['duration_second'] <0, 'end_time'] =  data.loc[data['duration_second'] <0,'end_time'] + pd.DateOffset(1)
data['duration_second'] = (data['end_time'] - data['start_time']).astype('timedelta64[s]')
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score


x = data.drop(['start_time', 'end_time', 'rating'], axis=1)
X = pd.get_dummies(x, columns = ["user_id","car_id","region"])
Y = data['rating'] 
#
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB

# clf = DecisionTreeClassifier(max_depth = 5)
x = data.drop(['start_time', 'end_time', 'rating'], axis=1)
X = pd.get_dummies(x, columns = ["user_id","car_id","region"])
Y = data['rating'] 

X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size = 0.33, random_state=42)

```

```python
gnb = GaussianNB()
gnb.fit = gnb.fit(X_train, y_train)

pred = gnb.fit.predict(X_test)
# np.mean(fited == Y)
accuracy_score(y_test,pred)*100
```

```python
clf = DecisionTreeClassifier(max_depth = 5)
clf = clf.fit(X, Y)
pred = clf.predict(X_test)
# np.mean(fited == Y)
accuracy_score(y_test,pred)*100
```

```python
from sklearn.externals.six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus
feature_names=X.columns

dot_data = StringIO()
export_graphviz(clf, out_file=dot_data,  
                feature_names=feature_names,
                filled=True, rounded=True,
                special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
# graph.write_png("spanish_test.png")
```
